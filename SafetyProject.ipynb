```python
import pandas as pd

# Load file
file_path = r"C:\Users\Danso Emmanuel Effah\Downloads\Telegram Desktop\Near_Miss__Dataset_200_Rows.xlsx"
df = pd.read_excel(file_path)

# STEP 1: Clean column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[^\w]", "", regex=True)
)

# STEP 2: Show columns (VERY IMPORTANT)
print("Columns after cleaning:")
print(df.columns.tolist())


# STEP 3: AUTO-FIX COLUMN NAMES (SMART MATCHING)

# Find columns that contain keywords
likelihood_col = [col for col in df.columns if "likelihood" in col]
consequence_col = [col for col in df.columns if "consequence" in col]
risk_col = [col for col in df.columns if "risk" in col and "score" in col]

# Rename them properly
if likelihood_col:
    df.rename(columns={likelihood_col[0]: "likelihood"}, inplace=True)

if consequence_col:
    df.rename(columns={consequence_col[0]: "consequence"}, inplace=True)

if risk_col:
    df.rename(columns={risk_col[0]: "risk_score"}, inplace=True)

# STEP 4: Confirm fix
print("\nUpdated Columns:")
print(df.columns.tolist())


# STEP 5: Now safely convert to numeric
df['likelihood'] = pd.to_numeric(df['likelihood'], errors='coerce')
df['consequence'] = pd.to_numeric(df['consequence'], errors='coerce')
df['risk_score'] = pd.to_numeric(df['risk_score'], errors='coerce')

print("\n✅ Columns fixed successfully!")
```

```python
import pandas as pd
from prophet import Prophet

# Load data
df = pd.read_csv("clean_near_miss.csv")

# Prepare data
df['date_reported'] = pd.to_datetime(df['date_reported'])

# Aggregate daily incidents
df_grouped = df.groupby('date_reported').size().reset_index(name='y')

# Rename for Prophet
df_grouped.columns = ['ds', 'y']

# Train model
model = Prophet()
model.fit(df_grouped)

# Create future dates until 2027
future = model.make_future_dataframe(periods=365*2)

# Predict
forecast = model.predict(future)

# Save results
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv("forecast_2027.csv", index=False)

print("✅ Forecast for 2027 created!")
```

```python
import pandas as pd
from prophet import Prophet

# Load data
df = pd.read_csv("clean_near_miss.csv")

# Prepare data
df['date_reported'] = pd.to_datetime(df['date_reported'])

# Aggregate
df_grouped = df.groupby('date_reported').size().reset_index(name='y')
df_grouped.columns = ['ds', 'y']

# Train model
model = Prophet()
model.fit(df_grouped)

# Future dates
future = model.make_future_dataframe(periods=365*2)

# Predict
forecast = model.predict(future)

# OPTIONAL: rename for your use
forecast_2027 = forecast

# Save
forecast_2027[['ds','yhat','yhat_lower','yhat_upper']].to_csv("forecast_2027.csv", index=False)

print("✅ Done successfully")
```
